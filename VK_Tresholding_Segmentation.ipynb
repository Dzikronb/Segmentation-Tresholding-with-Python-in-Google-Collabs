```python
import cv2 
import numpy as np 
from google.colab.patches import cv2_imshow
```

```python
image1 = cv2.imread('rambu.jpg')
cv2_imshow(image1)
```

```python
rambu_gray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(rambu_gray)
```

```python
ret, thresh1 = cv2.threshold(rambu_gray, 120, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(rambu_gray, 120, 255, cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(rambu_gray, 120, 255, cv2.THRESH_TRUNC)
ret, thresh4 = cv2.threshold(rambu_gray, 120, 255, cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(rambu_gray, 120, 255, cv2.THRESH_TOZERO_INV)
```

```python
cv2_imshow(thresh1)
cv2_imshow(thresh2)
cv2_imshow(thresh3)
cv2_imshow(thresh4)
cv2_imshow(thresh5)
```
